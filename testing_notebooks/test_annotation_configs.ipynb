```python
import os

os.environ["ARIZE_LOG_ENABLE"] = "true"
os.environ["ARIZE_LOG_LEVEL"] = "debug"
os.environ["ARIZE_LOG_STRUCTURED"] = "false"
```

```python
from arize import ArizeClient

client = ArizeClient() # API key from env var

print(client)
```

## List Annotation Configs

### List all the annotation configs the user has access to

```python
ac_list_resp = client.annotation_configs.list()
```

### Limit the number of annotation configs returned

```python
ac_list_resp = client.annotation_configs.list(limit=10)
```

### List the annotation configs the user has access to within a space

```python
ac_list_resp = client.annotation_configs.list(space_id=SPACE_ID)
```

### Access the annotation configs in the response

```python
ac_list_resp
```

### Format to dictionary

```python
ac_list_resp.to_dict()
```

### Format to dataframe (our own extension)

```python
ac_list_resp.to_df()
```

## Create Annotation Config

```python
resp = client.annotation_configs.create(
    name="Animals1",
    space_id=SPACE_ID,
    type="categorical",
    values=[{"label": "dog", "score": 1}, {"label": "cat", "score": 0}],
    optimization_direction="maximize",
)
resp
```

```python
resp.to_dict()
```

```python
AC_ID = resp.actual_instance.id
```

## Get Annotation Config by ID

```python
ac = client.annotation_configs.get(id=AC_ID)
ac
```

## Convert to dictionary

```python
ac.to_dict()
```

## Delete Annotation Config By ID

```python
ac_list_resp = client.annotation_configs.list(space_id=SPACE_ID, limit=5)
ac_list_resp.to_df()
```

```python
resp = client.annotation_configs.delete(id=AC_ID)
resp
```
