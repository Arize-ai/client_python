# Create Experiment

Select a dataset and copy 2 example IDs, example IDs MUST match

```python
DATASET_ID = "RGF0YXNldDozMjU0MDY6b3VqUA=="
ds = client.datasets.get(dataset_id=DATASET_ID)
ds
```

```python
examples = client.datasets.list_examples(dataset_id=DATASET_ID, limit=2)
example_ids = examples.to_df()["id"]
```

```python
from datetime import datetime
from arize.experiments.types import ExperimentTaskFieldNames


def append_timestamp(text: str) -> str:
    timestamp = datetime.now().strftime("%Y%m%d%H%M%S")
    return f"{text}_{timestamp}"


runs = [
    {
        "id": append_timestamp("id"),
        "ex_id": example_ids[0],
        "question": "Best food France or Spain?",
        "answer": "Spain",
        "extra-int": 2,
        "extra-str": "blabla",
    },
    {
        "id": append_timestamp("id"),
        "ex_id": example_ids[1],
        "question": "Best football player?",
        "answer": "Ronaldo",
        "extra-int": 5,
        "extra-str": "kekeke",
    },
]
task_fields = ExperimentTaskFieldNames(
    example_id="ex_id",
    output="answer",
)
```

```python
resp = client.experiments.create(
    name=append_timestamp("kiko-test-experiment"),
    dataset_id=DATASET_ID,
    experiment_runs=runs,
    task_fields=task_fields,
)
resp
```

# Get Experiment by ID

```python
EXPERIMENT_ID = resp.id
exp = client.experiments.get(experiment_id=EXPERIMENT_ID)
exp
```

## Convert to dictionery

```python
exp.to_dict()
```
